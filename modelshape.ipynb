```python
import timm

vs = timm.list_models('vit_s*224*')
vs
```

```python
import torch
for v in vs :
    model = timm.create_model(v,)
    p_n = sum(p.numel() for p in model.parameters())
    ip = torch.randn((1, 3, 224, 224))
    s = model.forward_features(ip).shape
    print(v, s, p_n)
```
